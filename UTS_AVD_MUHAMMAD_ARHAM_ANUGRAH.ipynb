## **Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['AQI Value'])
plt.title('Boxplot of AQI Value')
plt.ylabel('AQI Value')
plt.show()
```

Analisis outlier menunjukkan bahwa terdapat sejumlah nilai AQI yang berada jauh di atas mayoritas data lainnya. Nilai ekstrem tersebut mengindikasikan adanya kondisi kualitas udara yang sangat buruk pada beberapa negara atau waktu tertentu.

Keberadaan outlier pada data AQI merupakan hal yang penting untuk diperhatikan karena dapat mencerminkan kondisi pencemaran udara ekstrem akibat faktor seperti aktivitas industri, kepadatan kendaraan, kebakaran hutan, maupun faktor lingkungan lainnya.

Selain itu, outlier juga menyebabkan distribusi data menjadi tidak merata dan dapat memengaruhi nilai rata-rata keseluruhan dataset. Oleh sebab itu, analisis outlier memberikan gambaran bahwa kondisi kualitas udara antar negara memiliki perbedaan yang cukup signifikan.

# **EDA**

## **Comparison**

```python
# Mengelompokkan rata-rata AQI per negara
variabel = df.groupby('Country')['AQI Value'].mean().sort_values(ascending=False)


# Ambil top 10 tertinggi dan terendah
top10_worst = variabel.head(10)
top10_best  = variabel.tail(10).sort_values(ascending=True)


# Plot: 2 subplot berdampingan
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# --- Kiri: 10 negara AQI tertinggi ---
sns.barplot(x=top10_worst.values, y=top10_worst.index,
            hue=top10_worst.index, palette=['#e74c3c'] * 10,
            legend=False, ax=axes[0])


axes[0].set_title('Top 10 Negara AQI Tertinggi (Terburuk)', fontweight='bold')
axes[0].set_xlabel('Rata-rata AQI Value')
axes[0].set_ylabel('Negara')


# Label nilai di ujung bar
for i, val in enumerate(top10_worst.values):
    axes[0].text(val + 1, i, f'{val:.1f}', va='center', fontsize=10)


# --- Kanan: 10 negara AQI terendah ---
sns.barplot(x=top10_best.values, y=top10_best.index,
            hue=top10_best.index, palette=['#2ecc71'] * 10,
            legend=False, ax=axes[1])


axes[1].set_title('Top 10 Negara AQI Terendah (Terbersih)', fontweight='bold')
axes[1].set_xlabel('Rata-rata AQI Value')
axes[1].set_ylabel('Negara')


for i, val in enumerate(top10_best.values):
    axes[1].text(val + 0.3, i, f'{val:.1f}', va='center', fontsize=10)


plt.suptitle('Perbandingan AQI Berdasarkan Negara', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
```

Visualisasi bar chart memperlihatkan adanya perbedaan yang cukup signifikan antara negara dengan nilai AQI tertinggi dan negara dengan nilai AQI terendah.

Negara dengan AQI tertinggi menunjukkan tingkat pencemaran udara yang lebih buruk dibanding negara lainnya. Kondisi ini dapat dipengaruhi oleh berbagai faktor seperti industrialisasi, kepadatan penduduk, emisi kendaraan, serta aktivitas pembakaran.

Sebaliknya, negara dengan AQI terendah memiliki kualitas udara yang relatif lebih baik. Hal ini dapat mengindikasikan bahwa negara tersebut memiliki tingkat polusi yang lebih rendah atau sistem pengendalian lingkungan yang lebih baik.

Perbandingan antar negara pada visualisasi ini membantu mengidentifikasi negara-negara yang memerlukan perhatian lebih dalam pengelolaan kualitas udara dan pengendalian polusi.

## **Composition**

```python
top3 = (
    df.groupby("Country")["AQI Value"]
    .max()
    .sort_values(ascending=False)
    .head(3)
)

plt.figure(figsize=(7,7))
plt.pie(
    top3.values,
    labels=top3.index,
    autopct='%1.1f%%'
)

plt.title("Komposisi Top 3 Negara dengan AQI Tertinggi")
plt.show()
```

Pie chart menunjukkan komposisi kontribusi tiga negara dengan nilai AQI tertinggi terhadap total keseluruhan AQI pada kelompok tersebut.

Visualisasi memperlihatkan bahwa terdapat negara yang memiliki proporsi lebih dominan dibanding negara lainnya. Semakin besar persentase yang ditampilkan pada pie chart, maka semakin besar kontribusi negara tersebut terhadap tingkat pencemaran udara tertinggi.

Perbedaan proporsi antar negara menunjukkan bahwa tingkat pencemaran udara tidak tersebar secara merata. Beberapa negara memiliki tingkat AQI yang jauh lebih tinggi dibanding negara lain dalam kelompok top 3.

Insight ini memperlihatkan adanya ketimpangan kualitas udara antar negara dan menunjukkan bahwa beberapa wilayah memiliki permasalahan pencemaran udara yang lebih serius.

## **Distribution**

```python
plt.figure(figsize=(8,5))
plt.hist(df["AQI Value"], bins=20)

plt.title("Histogram Persebaran AQI Value")
plt.xlabel("AQI Value")
plt.ylabel("Frekuensi")
plt.show()
```

Histogram pada setiap kolom numerik menunjukkan pola distribusi data yang berbeda-beda. Sebagian besar histogram memperlihatkan bahwa data tidak terdistribusi secara normal atau simetris.

Pada kolom AQI Value, distribusi data cenderung terkonsentrasi pada rentang nilai tertentu dengan sejumlah kecil data yang memiliki nilai sangat tinggi. Kondisi ini menyebabkan distribusi menjadi miring ke kanan (right skewed).

Hal tersebut menunjukkan bahwa mayoritas pengamatan memiliki tingkat AQI pada kategori tertentu, namun terdapat beberapa kondisi ekstrem dengan kualitas udara yang jauh lebih buruk.

Selain itu, histogram juga membantu mengidentifikasi pola kepadatan data, rentang nilai dominan, serta kemungkinan keberadaan outlier pada data numerik.
